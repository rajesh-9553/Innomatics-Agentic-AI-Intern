# 1. Unique Words in a Sentence

# 2. Highest Salary from Employee Data

```python
# Create dict to store salary
employees={
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Set max salary to 0
max_salary=0
max_employee=""

for name,salary in employees.items():

  # Check if current salary is greater than max_salary
  if salary>max_salary:
    max_salary=salary
    max_employee=name

print("Highest Salary:", max_employee,"-",max_salary)
```

# 3. Find Maximum and Minimum Values

```python
# Take input from user
numbers = list(map(int, input().split(",")))

# Initialize both max and min with first element
max_value=numbers[0]
min_value=numbers[0]

for i in numbers:

  # Check maximum value
  if i>max_value:
    max_value=i

  # Check minimum value
  if i<min_value:
    min_value=i

print("List:",numbers)
print("Max:", max_value)
print("Min:", min_value)
```

# 4. Count Products Above a Price Threshold

```python
# Take prices as input
prices = list(map(int, input().split(",")))

# Set count to 0
count=0

for price in prices:

  # Check if price is greater than 1000
  if price>1000:
    count+=1  # Increase count if condition is true

print("Products above 1000:",count)
```

# 5. Calculate Attendance Percentage

```python
# Create a list representing attendance
attendance = ["P", "P", "A", "P", "P"]

# Set count to 0
count=0

for att in attendance:

  # If the student is present, increase count
  if att=="P":
    count+=1

# Calculate percentage
percentage=(count/len(attendance))*100

print("Attendance Percentage:",percentage)
```

# 7. Count Character Frequency

```python
# Take string input from user
text=input()

# Create an empty dict to store char freqs
freq={}

for ch in text:

  # Check if char in dict, if yes increment the count
  if ch in freq:
    freq[ch]+=1

  # If char doesn't exists, add it with count 1
  else:
    freq[ch]=1

print(freq)
```
