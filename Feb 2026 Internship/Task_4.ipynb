# 1. Mobile Recharge Validation System

```python
def validate_recharge(amount):
  valid_plans = [199, 299, 399, 599]

  if amount < 50:
    print("Minimum recharge amount is 50.")
    return False

  if amount not in valid_plans:
    print("Invalid plan selected.")
    print("Available plans are: 199, 299, 399, 599")
    return False

  return True

while True:
  try:
    recharge_amount = int(input())

    if validate_recharge(recharge_amount):
      print("Recharge successful!")
      break

    print("Please try again.\n")

  except ValueError:
    print("Please enter a valid number")
```

# 2. Inventory Reorder Alert System

```python
def check_inventory(stock_data):
  for product, quantity in stock_data.items():
    if quantity < 15:
        print(f" {product}: Reorder Alert (Stock: {quantity})")
    else:
        print(f" {product}: Stock Available (Stock: {quantity})")

inventory={
    "Rice": 10,
    "Wheat": 25,
    "Sugar": 8,
    "Oil": 20,
    "Salt": 14
}

check_inventory(inventory)
```

# 3. Student Result Processing System

```python
def process_result(marks):
  total = 0

  for mark in marks:
    total+=mark

  avg= total / len(marks)

  if avg >= 50:
    return avg, "Pass"
  else:
    return avg, "Fail"

student_marks = [60, 45, 70, 55, 40]
avg, res = process_result(student_marks)

print(f"Average Marks: {avg}")
print(f"Result: {res}")
```

# 4. Cab Fare Estimator with Retry Option

```python
def calculate_fare(dist, is_peak):
  base_fare = 50
  per_km_rate = 12

  fare = base_fare + (dist * per_km_rate)

  if is_peak:
    fare += fare * 0.25

  return fare

while True:
    try:
      dist = float(input())

      if dist <= 0:
        print("Distance must be greater than 0.\n")
        continue
      peak_input = input("Is it peak hour? (yes/no): ").lower()

      if peak_input not in ["yes", "no"]:
        print(" Please enter 'yes' or 'no'.\n")
        continue

      is_peak = True if peak_input == "yes" else False

      total_fare = calculate_fare(dist, is_peak)

      print(f"\n Total Fare: ₹{total_fare:.2f}")
      print("Booking Confirmed!")
      break

    except:
      print("Please enter a valid distance.\n")

```

# 5. Employee Attendance Eligibility Checker


```python
def check_attendance(attendance_list):
  total_days = len(attendance_list)
  present_days = 0

  for day in attendance_list:
    if day == "P":
      present_days += 1

  attendence_percentage = (present_days / total_days) * 100

  if attendence_percentage >= 75:
    return attendence_percentage, "Eligible"
  else:
    return attendence_percentage, "Not eligible"

attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]

percentage, status = check_attendance(attendance)

print("Attendance Percentage:", percentage)
print("Status:", status)
```

# 6. Password Strength Checker

```python
def check_password_strength(password):
  if len(password) < 8:
    return "Weak Password (Minimum 8 characters required)"

  has_digit = False
  has_special = False
  special_chars = "@#$"

  for char in password:
    if char.isdigit():
      has_digit = True
    if char in special_chars:
      has_special = True

  if has_digit and has_special:
    return "Strong Password"
  else:
    return "Weak Password(Must contain at least one digit and one special character @#$)"

pwd = input()
res = check_password_strength(pwd)
print(res)
```
