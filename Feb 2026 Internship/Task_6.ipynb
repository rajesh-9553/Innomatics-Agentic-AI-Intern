# 1. Smart Parking Lot Management System

```python
def smart_parking(capacity, logs):
  curr=0
  peak=0

  for log in logs:
    if log=='IN':
      curr+=1
    elif log=='OUT':
      if curr>0:
        curr-=1
    if curr>peak:
      peak=curr
  print("Current Parked Vehicles:", curr)
  print("Peak Parking Usage:", peak)

  if curr>capacity:
      print("ALERT: Parking capacity exceeded!")
  else:
      print("Parking Status: Available")

capacity=50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
smart_parking(capacity, logs)
```

# 2. Online Food Delivery Time Estimator

```python
def estimate_delivery_time(distance, traffic, weather):
  base_speed=30
  base_time=(distance/base_speed)*60

  if traffic =="Low":
    traffic_delay=5
  elif traffic=="Medium":
    traffic_delay=10
  elif traffic=="High":
    traffic_delay=20
  else:
    traffic_delay=0

  if weather=="Clear":
    weather_delay=0
  elif weather=="Rainy":
    weather_delay=15
  elif weather=="Cloudy":
    weather_delay=5
  elif weather=="Strom":
    weather_delay=25
  else:
    weather_delay=0

  final_time=base_time+traffic_delay+weather_delay
  print("Estimated Delivery Time:", round(final_time), "minutes")

estimate_delivery_time(8, "High", "Rainy")
```

# 3.Movie Theatre Seat Occupancy

```python
def theatre_occupancy(total_seats, booked_seats):
  booked_count=len(booked_seats)

  occupancy=(booked_count/total_seats)*100

  print("Occupancy:", round(occupancy),"%")

  if occupancy==100:
    print("Show Status: Housefull")
  elif occupancy>=70:
    print("Show Status: Almost Full")
  else:
    print("Show Status: Available")

  if occupancy>=90:
    print("Suggestion: Consider opening additional show")

total=200
booked=[1]*150
theatre_occupancy(total, booked)
```

# 4. Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
  total=0

  for usage in cpu_readings:
    total+=usage
  avg=total/len(cpu_readings)
  print("Average CPU Usage:", round(avg),"%")

  if avg<50:
    print("Server Status: Normal")
  elif avg<=80:
    print("Server Status: Warning")
  else:
    print("Server Status: Critical")


readings = [45, 60, 70, 85, 90]
classify_server_load(readings)
```

# 5. Smart Classroom Resource Usage Monitor

```python
def classroom_resource_monitor(resource_usage):
  threshold=8
  overused=[]

  for resource, hours in resource_usage.items():
    if hours>threshold:
      overused.append(resource)

  if overused:
      print("Overused Resources:", ", ".join(overused))
      print("Energy Alert: Yes")
  else:
      print("Overused Resources: None")
      print("Energy Alert: No")

usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_resource_monitor(usage)
```

# 6. Online Event Registration Capacity Controller

```python
def event_registration_controller(capacity, total_registration):
  confirmed=0
  waitlist=0

  for i in range(total_registration):
    if confirmed<capacity:
      confirmed+=1
    else:
      waitlist+=1

  print("Cofirmed Registeration:", confirmed)
  print("Waitlisted Users:", waitlist)

  if confirmed==capacity:
    print("Event Status: Full")
  else:
    print("Event Status: Available")

  if confirmed>=capacity:
      print("Registration Status: Closed")
  else:
      print("Registration Status: Open")

event_registration_controller(100,105)
```
