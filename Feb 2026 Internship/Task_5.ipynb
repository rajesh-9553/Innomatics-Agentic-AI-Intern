# 1. Social Media – Post Engagement Analyzer

```python
def post_engagement_analyzer(likes_list):
  total_likes=0

  for likes in likes_list:
    total_likes+=likes

  print("Total Liked:", total_likes)

  if total_likes >=1000:
    print("Post Status: Viral Post")
  else:
    print("Post Status: Normal Engagement")

likes=list(map(int, input().split(",")))
post_engagement_analyzer(likes)
```

# 3. Agriculture – Rainfall Adequacy Checker

```python
def rainfall_checker(rainfall_list, required_level):
  total=0

  for rain in rainfall_list:
    total+=rain

  average=total/len(rainfall_list)
  print("Average Rainfall:", int(average))

  if average>= required_level:
    print("Rainfall Status: Adequate Rainfall")
  else:
    print("Rainfall Status: Inadequate Rainfall")

rain_data=list(map(int, input().split(",")))
required=int(input())
rainfall_checker(rain_data,required)
```

# 4. Social Media – Duplicate Account Detection

```python
def duplicate_account_detection(usernames):
  unique_users=set(usernames)

  if len(unique_users)!=len(usernames):
    print("Duplicate Accounts Found: Yes")
  else:
    print("Duplicate Acoounts Found: No")

users=list(map(str, input().split(",")))
duplicate_account_detection(users)
```

# 6. Agriculture – Premium Crop Price Filter

```python
def premium_crop_filter(price_list):
  premium_crops=[]

  for price in price_list:
    if price>2000:
      premium_crops.append(price)

  print("Premium Crops:", premium_crops)

crop_prices=list(map(int, input().split(",")))
premium_crop_filter(crop_prices)
```

# 7. System Monitoring – Application Health

```python
def application_health_checker(errors):
  print("Error Count:", errors)

  if errors==0:
    print("System Status: Healthy")
  elif errors<=5:
    print("System Status: Minor Issues")
  else:
    print("System Status: Critical Issues")

errors=int(input())
application_health_checker(errors)
```

# 9. E-Learning – Student Attendance Eligibility System

```python
def attendance_eligibility(attendance_list):
  total_days=len(attendance_list)
  present_days=0

  for status in attendance_list:
    if status=='P':
      present_days+=1

  percentage=(present_days/total_days)*100
  print("Attendance Percentage:", percentage)

  if percentage>=75:
    print("Exam Eligibility: Eligible")
  else:
    print("Exam Eligibility: Not Eligible")

attendance=["P", "P", "A", "P", "P"]
attendance_eligibility(attendance)
```
