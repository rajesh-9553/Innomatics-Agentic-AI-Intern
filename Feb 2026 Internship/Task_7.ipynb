# 1. Smart Light Controller

```python
class SmartLight:

  def __init__(self, name):
    self.name=name
    self.status="OFF"

  def turn_on(self):
    self.status="ON"
    print(f"{self.name} is {self.status}")

  def turn_off(self):
    self.status="OFF"
    print(f"{self.name} is {self.status}")

  def show_status(self):
    print(f"{self.name} is {self.status}")

light_name="Bedroom Light"
action="ON"
light=SmartLight(light_name)

if action=="ON":
  light.turn_on()
elif action=="OFF":
  light.turn_off()
else:
  print("Invalid Action")
```

# 2. Employee ID Card System

```python
class Employee:

  def __init__(self, name, emp_id, dept):
    self.name=name
    self.emp_id=emp_id
    self.dept=dept

  def display_id_card(self):
    print("Employee ID Card")
    print(f"Name: {self.name}")
    print(f"ID: {self.emp_id}")
    print(f"Dept: {self.dept}")

name="Rahul"
emp_id="EMP103"
dept="IT"

emp=Employee(name, emp_id, dept)
emp.display_id_card()
```

# 3. Mobile Contact Record

```python
class Contact:

  def __init__(self, name, phone):
    self.name=name
    self.phone=phone

  def display_contact(self):
    print("Contact Saved")
    print(f"Name: {self.name}")
    print(f"Phone: {self.phone}")

contact_name="Anita"
contact_phone="9876543210"

contact=Contact(contact_name, contact_phone)
contact.display_contact()
```

# 4. Product Price Tag Generator

```python
class Product:

  def __init__(self, name, price):
    self.name=name
    self.price=price

  def print_price_tag(self):
    print(f"Name: {self.name}")
    print(f"Price: {self.price}")


product_name="Laptop"
price=100000

product=Product(product_name, price)
product.print_price_tag()
```

# 5. Movie Rating Display System

```python
class Movie:

  def __init__(self, name, rating):
    self.name=name
    self.rating=rating

  def display_movie(self):
    print(f"Name: {self.name}")
    print(f"Rating: {self.rating}/5")

movie_name="Inception"
rating=4.8

movie1=Movie(movie_name, rating)
movie1.display_movie()
```

# 6. Delivery Address Manager

```python
class Delivery:

  def __init__(self, cust_name, address):
    self.cust_name=cust_name
    self.address=address

  def print_delivery_address(self):
    print("Delivery Details")
    print(f"Customer Name: {self.cust_name}")
    print(f"Address: {self.address}")

cust_name="Suman"
address="Hyderabad"

delivery1=Delivery(cust_name, address)
delivery1.print_delivery_address()
```
